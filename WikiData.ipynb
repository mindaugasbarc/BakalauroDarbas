```
sentences = []

with open("/gdrive/My Drive/data/ltwikidata.txt", "r") as wiki_file:
  sentences = wiki_file.readlines()

sentences = [s for index,s in enumerate(sentences) if len(s) > 50 and len(s) < 150]
```

```
with open("/gdrive/My Drive/data/ltwikidata_filtered.txt", "w") as wiki_file:
  wiki_file.writelines(sentences)
```
