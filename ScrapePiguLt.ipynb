```
from urllib.request import urlopen, Request
```

```
pigu_url = "https://pigu.lt/lt/baldai-baldai-internetu-ir-interjeras"
```

```
def get_parser(page_url):
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
  request = Request(pigu_url, headers=headers)
  webpage = urlopen(request)
  page_html = webpage.read()
  webpage_parser = BeautifulSoup(page_html, "html.parser")
  return webpage_parser
```

```
import random

def get_category_links(page_url):
  parser = get_parser(page_url)
  links = []
  for menu in parser.find_all('ul', class_='menu'):
    for a in menu.find_all('a', href=True):
      link = a['href']
      if ('http' in link):
        links.append(link)
  random.shuffle(links)
  return links

def get_product_links(page_url):
  parser = get_parser(page_url)
  links = set()
  for main_block in parser.find_all('div', class_='product-list'):
    print("main block found:")
    print(main_block)
    for a in main_block.find_all('a', href=True):
      link = a['href']
      if ("?id=" in link):
        links.add('https:/pigu.lt' + link)  
  return links

def get_product_links_2(page_url):
  parser = get_parser(page_url)
  links = set()
  for a in parser.find_all('a', class_='image-wrap'):
    link = a['href']
    if ("?id=" in link):
      links.add(link)
  return links
```

```
import random

def get_all_product_links(links, layer, links_traversed = set()):
  print("links_traversed")
  print(links_traversed)
  product_links = set()
  category_links = set()
  if (layer > 3):
    return product_links

  for link in links:
    gotten_product_links = get_product_links(link)
    print("gotten product links")
    print(gotten_product_links)
    product_links.update(gotten_product_links)
    gotten_category_links = get_category_links(link)
    print("gotten category links")
    print(list(gotten_category_links)[50:75])
    category_links.update(gotten_category_links)

  links_traversed.update(links)
  product_links.update(get_all_product_links([l for l in list(category_links)[50:75] if l not in links_traversed], layer + 1, links_traversed))
  return product_links
```

```
def get_web_driver():
  chrome_options = webdriver.ChromeOptions()
  chrome_options.add_argument('--headless')
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')
  return webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```
class Review:
  def __init__(self, rating, description):
    self.rating = rating
    self.description = description

  def __str__(self):
    return ({ "rating": self.rating,
              "description": self.description}).__str__()
    
  def __repr__(self):
    return self.__str__()
    
def get_reviews(driver : webdriver, link : str):
  driver.get(link)
  driver.execute_script("window.scrollTo(0, 10000)")
  review_blocks = driver.find_elements_by_class_name("one_item")
  reviews = []
  for review_block in review_blocks:
    ratings = review_block.find_elements_by_class_name("rating_title")
    descriptions = review_block.find_elements_by_class_name("q_text")

    if (len(ratings) == 1 and len(descriptions) == 1):
      reviews.append(Review(ratings[0].text[0], descriptions[0].text))
  return reviews
```

```
def get_product_links_selenium(category : str):
  driver = get_web_driver()
  driver.get(category)
  driver.execute_script("window.scrollTo(0, 10000)")
  image_a_tags = driver.find_elements_by_class_name("image-wrap")
  product_links = []
  for a in image_a_tags:
      product_link = a.get_attribute("href")
      if (product_link == None):
        continue
      if ("pigu.lt" in product_link and "id" in product_link):
        product_links.append(product_link)
  return product_links
```

```
def get_products_selenium(category_links:list):
  products = []
  for category_link in category_links:
    products.extend(get_product_links_selenium(category_link))
  return products

def get_product_reviews(product_links:list):
  reviews = []
  driver = get_web_driver()
  for product_link in product_links:
    reviews.extend(get_reviews(driver, product_link))
  return reviews
```

```
import random
def find_products(page_url):
  driver = get_web_driver()
  product_links = set()
  categories = get_category_links(page_url)
  for index,category in enumerate(categories):
    if (index > 50):
      return product_links
    products = get_product_links_selenium(category)
    if (len(products) == 0):
      products = get_products_internal(category)
      product_links.update(products)
    else:
      product_links.update(products)
  return product_links

def get_products_internal(page_url):
  categories = get_category_links(page_url)
  for category in categories:
    products = get_product_links_selenium(category)
    if (len(products) == 0):
      return get_products_internal(category)
    else:
      return products
  return []    
```
